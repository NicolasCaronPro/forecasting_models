```python
start_date = "2023-01-01"
end_date = "2023-12-31"
DATA_DIR = Path('../data/features/')

etab = ETAB_NAMES[0][1]

date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# loc = Location(etab)
# features = [HopitalFeatures(name='hopitalfeatures', include_hnfc_moving=False, include_nb_hospit=False)]
# df = features[0].fetch_data_function(location=loc, feature_dir=DATA_DIR / features[0].name, start_date=start_date, stop_date=end_date)
```

```python


# Sample DataFrame with Date index and Visitors column

data = {
    'Date': date_range,
    'Total': [200, 210, 220, 230, 240] * int(len(date_range)/5)  # Example visitor counts
}
df = pd.DataFrame(data).set_index('Date')

# Extract month from the Date index
df['Month'] = df.index.month
df['Day_of_Week'] = df.index.weekday  # 0 = Monday, 6 = Sunday
df['Year'] = df.index.year
df['Day_of_Month'] = df.index.day
df['Quarter'] = df.index.quarter
df['Is_Weekend'] = df.index.weekday >= 5
df['Week_of_Year'] = df.index.isocalendar().week
df['Is_Month_Start'] = df.index.is_month_start
df['Is_Month_End'] = df.index.is_month_end
df['Day_of_Year'] = df.index.dayofyear
df['Is_Year_Start'] = df.index.is_year_start
df['Is_Year_End'] = df.index.is_year_end



# Calculate rolling mean over the last 31 days, excluding the current day
df['Rolling_31_day_mean'] = df['Total'].shift(1).rolling(window=31, min_periods=1).mean()
df['Rolling_7_day_mean'] = df['Total'].shift(1).rolling(window=7, min_periods=1).mean()
df['Rolling_365_day_mean'] = df['Total'].shift(1).rolling(window=365, min_periods=1).mean()

# print(df[['Total', 'Rolling_31_day_mean']])
```
